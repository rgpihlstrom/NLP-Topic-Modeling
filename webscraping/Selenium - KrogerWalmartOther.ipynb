```python
def checkifThereisavalue(element):
    if len(element):
        result = element[0].text
    else:
        result = "Nothing"
    return result

def checkifThereisavalue1(element):
    if len(element):
        result = element[0].get_attribute("innerHTML")
    else:
        result = "Nothing"
    return result


def checkifThereisanImage(element):
   
    if len(element):
        images_ls = []
        for availableImages in element:
            images_ls.append(availableImages.get_attribute('src'))
        result = images_ls
    else:
        result = "Nothing"
    print("inside img list")    
    return result

def checkifThereisanImageKroger(element):
   
    if len(element):
        images_ls = []
        for availableImages in element:
            images_ls.append(availableImages.get_attribute('content'))
        result = images_ls
    else:
        result = "Nothing"
    print("inside img list")    
    return result
```

# Walmart

```python
df_urlFile = pd.read_csv("C:\\Users\\rgpih\\Documents\\WebScraping\\scrapy\\projects\\amazon2\\AmazonUrls.csv")
df_walmarturls = df_urlFile[df_urlFile["Target"]=="Walmart"]["URL"]


# url = 'https://www.walmart.com/reviews/product/957733881'
df=pd.DataFrame(columns=['reviewname','reviewTitles','postDate','starRating', 'reviewBody','verified','helpful','page','Product','Source','UPC','reviewerProfilerUrl','images'])

for Producturl in df_walmarturls:
    
    if Producturl:
        Cururl = Producturl
    else:
        Cururl = Cururl
        
    while Cururl != "Stop":
        browser.get(Cururl)
        time.sleep(5)

        reviews = browser.find_elements_by_xpath("//div[@class='Grid ReviewList-content']")
#         print(f'Number of Reviews: {len(reviews)}')

        for review in reviews:
            data={}
            starContainer = review.find_element_by_xpath(".//span[@class='stars-container']")
            #print(f' Number of star containers: {len(starContainer)}')

            fullStarsCount = len(starContainer.find_elements_by_xpath(".//span[@class='elc-icon star star-small star-rated elc-icon-star-rating']"))
            #print(f' Number of full stars: {(fullStarsCount)}')

            emptyStarsCount = len(starContainer.find_elements_by_xpath(".//span[@class='elc-icon star star-small star-empty elc-icon-star-rating']"))
            #print(f' Number of empty stars: {(emptyStarsCount)}')

            partialStarsCount = len(starContainer.find_elements_by_xpath(".//span[@class='elc-icon star star-medium star-partial elc-icon-star-rating']"))
            #print(f' Number of partial stars: {(partialStarsCount)}')
            numberOfStars = fullStarsCount+partialStarsCount

            data['reviewTitles'] = checkifThereisavalue(review.find_elements_by_xpath('.//h3[@itemprop="name"]'))
            data['starRating'] = numberOfStars
            data['postDate'] = checkifThereisavalue(review.find_elements_by_xpath('.//span[@itemprop="datePublished"]'))
            data['reviewBody'] = checkifThereisavalue(review.find_elements_by_xpath('.//div[@class="review-text"]/p'))
            data['images'] = checkifThereisanImage(review.find_elements_by_xpath('.//img[@class="review-media-thumbnail"]'))
            data['reviewname'] = checkifThereisavalue(review.find_elements_by_xpath('.//span[@itemprop="author"]'))
            data['helpful'] = checkifThereisavalue(review.find_elements_by_xpath('.//span[@class="yes-no-count underline"]'))
            data['Product'] = checkifThereisavalue(browser.find_elements_by_xpath('//h2[@class="prod-ProductTitle no-margin truncated wyr-product-title font-normal hide-content-max-s heading-b"]'))
            data['page'] = browser.current_url
            data['source'] = "https://www.walmart.com/reviews/product"
            data['reviewerProfilerUrl'] = "Nothing"
            df=df.append(data, ignore_index=True)
        nexturl = checkifThereisavalue(browser.find_elements_by_xpath("//button[@class='paginator-btn paginator-btn-next']"))
        if nexturl != "Nothing":
            browser.find_elements_by_xpath("//button[@class='paginator-btn paginator-btn-next']")[0].click()
            Cururl = browser.current_url
        else:
            Cururl = "Stop"                            
      
```
